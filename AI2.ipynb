# 1.3.1 Задание
# Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и единицами.

```python
# Сначала импортируем NumPy – это библиотека для работы с массивами и матрицами, очень полезная для численных операций.
import numpy as np

# Создаем пустую матрицу 8x8, заполненную нулями. Тип данных указываем как 'int', чтобы работали только целые числа.
Z = np.zeros((8,8), dtype = int)

# Теперь заполняем ее 'единицами' в шахматном порядке. Эта строка ставит '1' в нечетных строках (начиная со второй, т.е. с индекса 1)
# и четных столбцах (начиная с первого, т.е. с индекса 0).
Z[1::2, ::2] = 1

# А эта строка завершает шахматный узор, ставя '1' в четных строках (начиная с первой, т.е. с индекса 0)
# и нечетных столбцах (начиная со второго, т.е. с индекса 1).
Z[::2, 1::2] = 1

# И, наконец, выводим получившуюся красивую матрицу на экран.
print(Z)
```

# 1.3.2 Задание
 Создать 5x5 матрицу со значениями в строках от 0 до 4. Для создания необходимо использовать функцию arrange.

```python
# Подключаем библиотеку NumPy для удобной работы с массивами.
import numpy as np

# Создаем матрицу размером 5x5, которая пока заполнена нулями. Указываем тип данных как целые числа.
Z = np.zeros((5,5), dtype = int)

# Здесь хитрость: мы генерируем массив [0, 1, 2, 3, 4] с помощью np.arange(5).
# Затем NumPy автоматически добавляет этот массив к каждой строке нашей матрицы Z.
# Это называется 'broadcasting' – очень удобно!
Z += np.arange(5)

# Выводим готовую матрицу, где каждая строка начинается с 0 и заканчивается 4.
print(Z)
```

# 1.3.4 Задание
Создать матрицу с 0 внутри, и 1 на границах.

```python
# Снова импортируем NumPy для создания массивов.
import numpy as np

# Создаем квадратную матрицу 5x5, которая полностью заполнена единицами. Тип данных - целые числа.
Z = np.ones((5,5), dtype=int)

# Теперь самое интересное: делаем 'рамку'. Мы берем все элементы внутри (все строки, кроме первой и последней, и все столбцы, кроме первого и последнего)
# и присваиваем им значение 0. Таким образом, по краям остаются единицы, а внутри – нули.
Z[1:-1:,1:-1:] = 0

# Показываем, что получилось.
print(Z)
```

# 1.3.5 Задание
 Создать массив и отсортировать его в порядке убывания.

```python
# Подключаем NumPy, чтобы работать с массивами.
import numpy as np

# Создаем одномерный массив Z, который содержит числа от 0 до 13.
Z = np.array(range(14))

# Сортируем элементы массива Z по возрастанию. (В данном случае он уже отсортирован, так что эта строка ничего не меняет).
Z.sort()

# А вот здесь мы 'переворачиваем' массив, чтобы он был отсортирован по убыванию. Срез [::-1] создает обратную копию массива.
Z = Z[::-1]

# Выводим наш массив, отсортированный от большего к меньшему.
print(Z)
```

# 1.3.6 Задание
 Создать матрицу 5x5 со случайными значениями. Вывести форму, размер и размерность матрицы.

# 2.3.1 Задание
Задайте два Series, представляющих две точки в n-мерном пространстве. Найдите расстояние между ними.

```python
# Импортируем Pandas – это мощная библиотека для анализа данных, она умеет работать с таблицами (DataFrame) и одномерными рядами данных (Series).
import pandas as pd

# Представьте, что у нас есть две 'точки' в пространстве. Каждая точка представлена как Series в Pandas.
# Первая точка: [1, 2, 3, 4]
# Вторая точка: [5, 6, 7, 8]

# Мы хотим найти расстояние между этими двумя точками (это так называемое евклидово расстояние).
# Делаем это по шагам:
# 1. Вычитаем координаты одной точки из другой: (1-5), (2-6), (3-7), (4-8) = [-4, -4, -4, -4]
# 2. Каждый результат возводим в квадрат: (-4)**2 = 16. Получаем: [16, 16, 16, 16]
# 3. Суммируем все квадраты: 16 + 16 + 16 + 16 = 64
# 4. Извлекаем квадратный корень из этой суммы: корень(64) = 8.0
print(sum((pd.Series([1, 2, 3, 4])-pd.Series([5, 6, 7, 8]))**2)**.5)
```

# 3.3.2 Задание
Загрузить фрейм данных Iris.
Выполните нормализацию первого признака (sepal_length_cm) с помощью MinMaxScaler
и второго признака (sepal_width_cm) с помощью StandardScaler.


```python
# Подключаем NumPy для общих числовых операций.
import numpy as np
# Импортируем Pandas для работы с таблицами (DataFrame).
import pandas as pd
# Берем нужные инструменты для нормализации данных из библиотеки scikit-learn.
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Указываем ссылку на наш набор данных Iris.
url = "https://raw.githubusercontent.com/akmand/datasets/master/iris.csv"
# Загружаем данные из CSV-файла в DataFrame.
dataframe = pd.read_csv(url)


# Создаем 'нормализатор' MinMaxScaler. Он будет масштабировать значения в колонке так, чтобы они лежали в диапазоне от 0 до 1.
scalerM = MinMaxScaler()
# Применяем этот нормализатор к колонке 'sepal_length_cm'.
# Метод fit_transform сначала 'учится' на данных (находит минимум и максимум), а затем применяет масштабирование.
dataframe[['sepal_length_cm']] = scalerM.fit_transform(dataframe[['sepal_length_cm']])


# Теперь создаем другой 'стандартизатор' – StandardScaler. Он делает так, чтобы среднее значение в колонке стало 0, а стандартное отклонение – 1.
scalerZ = StandardScaler()
# Применяем его к колонке 'sepal_width_cm'.
# Аналогично, fit_transform сначала находит среднее и стандартное отклонение, а затем стандартизирует данные.
dataframe[['sepal_width_cm']] = scalerZ.fit_transform(dataframe[['sepal_width_cm']])

# Выводим обновленный DataFrame, где две колонки теперь нормализованы/стандартизированы.
dataframe
```
