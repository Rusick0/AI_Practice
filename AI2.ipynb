```python
# 1.3.1 Задание
# Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и единицами.
import numpy as np
Z = np.zeros((8,8), dtype = int)
Z[1::2, ::2] = 1
Z[::2, 1::2] = 1
print(Z)
```

```python
# 1.3.2 Задание
# Создать 5x5 матрицу со значениями в строках от 0 до 4.
# Для создания необходимо использовать функцию arrange.
import numpy as np
Z = np.zeros((5,5), dtype = int)
Z += np.arange(5)
print(Z)
```

```python
# 1.3.4 Задание
# Создать матрицу с 0 внутри, и 1 на границах.
import numpy as np
Z = np.ones((5,5), dtype=int)
Z[1:-1:,1:-1:] = 0
print(Z)
```

```python
# 1.3.5 Задание
# Создать массив и отсортировать его в порядке убывания.
import numpy as np
Z = np.array(range(14))
Z.sort()
Z = Z[::-1]
print(Z)
```

```python
# 1.3.6 Задание
# Создать матрицу 5x5 со случайными значениями.
# Вывести форму, размер и размерность матрицы.
import numpy as np
Z = np.random.random((5,5))
print("Форма матрицы:")
print(Z.shape)
print("Размер матрицы:")
print(Z.size)
print("Размерность матрицы:")
print(Z.ndim)
```

```python
# 3.3.2 Задание
# Загрузить фрейм данных Iris.
# Выполните нормализацию первого признака (sepal_length_cm) с помощью MinMaxScaler
# и второго признака (sepal_width_cm) с помощью StandardScaler.
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

url = "https://raw.githubusercontent.com/akmand/datasets/master/iris.csv"
dataframe = pd.read_csv(url)


scalerM = MinMaxScaler()
dataframe[['sepal_length_cm']] = scalerM.fit_transform(dataframe[['sepal_length_cm']])


scalerZ = StandardScaler()
dataframe[['sepal_width_cm']] = scalerZ.fit_transform(dataframe[['sepal_width_cm']])

dataframe
```
