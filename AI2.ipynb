# 1.3.1 Задание
# Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и единицами.

```python
import numpy as np

# Создаем матрицу 8x8 из нулей
Z = np.zeros((8,8), dtype=int)

# Заполняем в шахматном порядке: 1 в нечетных строках / четных столбцах
Z[1::2, ::2] = 1

# Заполняем в шахматном порядке: 1 в четных строках / нечетных столбцах
Z[::2, 1::2] = 1

print(Z)
```

# 1.3.2 Задание
 Создать 5x5 матрицу со значениями в строках от 0 до 4. Для создания необходимо использовать функцию arrange.

```python
import numpy as np

# Инициализируем матрицу 5x5 нулями
Z = np.zeros((5,5), dtype=int)

# Используем broadcasting для заполнения строк значениями от 0 до 4
Z += np.arange(5)

print(Z)
```

# 1.3.4 Задание
Создать матрицу с 0 внутри, и 1 на границах.

```python
import numpy as np

# Создаем матрицу 5x5 из единиц
Z = np.ones((5,5), dtype=int)

# Заполняем внутреннюю область нулями, оставляя границы
Z[1:-1, 1:-1] = 0

print(Z)
```

# 2.3.1 Задание
Найдитн евклидово расстояние между двумя Series (точками) a и b, не
используя встроенную формулу.


```python
import pandas as pd

# Точки в виде объектов Series
a = pd.Series([1, 2, 3, 4])
b = pd.Series([5, 6, 7, 8])

# Расчет евклидова расстояния через сумму квадратов разностей
distance = sum((a - b)**2)**.5

print(f"Евклидово расстояние: {distance}")
```

# 3.3.2 Задание
Загрузить фрейм данных по ссылке:
https://raw.githubusercontent.com/akmand/datasets/master/iris.csv.
Необходимо выполнить нормализацию первого числового признака
(sepal_length_cm) с использованием минимаксного преобразования, а
второго (sepal_width_cm) с задействованием z-масштабирования


```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler
#Разница между минмакс методом и маштабированием
url = "https://raw.githubusercontent.com/akmand/datasets/master/iris.csv"
dataframe = pd.read_csv(url)

# Min-Max нормализация (диапазон [0, 1]) для sepal_length_cm
scalerM = MinMaxScaler()
dataframe[['sepal_length_cm']] = scalerM.fit_transform(dataframe[['sepal_length_cm']])

# Z-масштабирование (mean=0, std=1) для sepal_width_cm
scalerZ = StandardScaler()
dataframe[['sepal_width_cm']] = scalerZ.fit_transform(dataframe[['sepal_width_cm']])

display(dataframe.head())
```
