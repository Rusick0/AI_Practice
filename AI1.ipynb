4.3.1 Задание

Задайте  массив случайных значений из интервала(0; 1).  
Рассчитайте средние и медианные значения для массива, сравните
результаты, какие выводы можно сделать о значениях?  
Постройте точечную диаграмму рассения полученного ряда.


```python
import random  # генерация случайных чисел
from statistics import median  # медиана
from matplotlib import pyplot as plt  # графики

n = int(input("Введите размер массива случайных чисел (целое число): "))

random_arr = []
for _ in range(n):
    random_arr.append(random.random())  # random() даёт число из интервала [0; 1)

avg_value = sum(random_arr) / len(random_arr)  # среднее значение
median_value = median(random_arr)  # медиана

print(f"Среднее значение массива: {avg_value:.3f}")
print(f"Медианное значение массива: {median_value:.3f}")

# --- Блок сравнения и выводов ---
diff = abs(avg_value - median_value)
print(f"\nРазница между средним и медианой: {diff:.4f}")
if diff < 0.1:
    print("Вывод: Среднее и медиана близки, что подтверждает отсутствие сильных отклонений и равномерность распределения.")
else:
    print("Вывод: Наблюдается различие значений, что часто случается при малом объеме выборки.")

# Точечная диаграмма рассеяния (индекс -> значение)
plt.grid(True)  # включаем сетку
plt.scatter(range(n), random_arr)  # точки: номер элемента -> значение
plt.title("Случайные значения из интервала (0; 1)")
plt.xlabel("Индекс элемента массива")
plt.ylabel("Значение")
plt.show()
```

4.3.2 Задание

Дана функция $ \frac{\sqrt{1 + e^{sqrt(x)} + cos(x^2)}}{|1 - sin^3(x)|} + ln|2x| $
Создать массив из 10 значений функции (𝑥, например, изменяется от 1 до 10). Выделить срез первой
половины массива и построить графики для основного массива –
линейный и для среза – точечный

```python
import numpy as np  # массивы и мат. функции
import matplotlib.pyplot as plt  # графики

# x меняется от 1 до 10 (10 значений)
x = np.arange(1, 11, 1)  # [1,2,...,10]

# y = sqrt(1 + e^{sqrt(x)} + cos(x^2)) / |1 - sin^3(x)| + ln|2x|
y = (np.sqrt(1 + np.exp(np.sqrt(x)) + np.cos(x**2)) / np.abs(1 - (np.sin(x))**3)) + np.log(np.abs(2 * x))

# Срез первой половины массива
half = len(x) // 2
x_slice = x[:half]
y_slice = y[:half]

form_y = ", ".join([f"{x:.3f}" for x in y])
form_y_slice = ", ".join([f"{x:.3f}" for x in y_slice])
print(f"Основной массив y (10 значений функции): [{form_y}]")
print(f"Срез первой половины массива y: [{form_y_slice}]")


# Линейный график для основного массива
plt.figure(figsize=(8, 6))
plt.grid(True)  # сетка
plt.plot(x, y)  # линейный график
plt.title("Линейный график функции (x от 1 до 10)")
plt.xlabel("Значения по x")
plt.ylabel("значения по y")
plt.show()

# Точечный график для среза
plt.figure(figsize=(8, 6))
plt.grid(True)  # сетка
plt.scatter(x_slice, y_slice)  # точечный график
plt.title("Точечный график (первая половина массива)")
plt.xlabel("Значения по x (срез)")
plt.ylabel("Значения по y (срез)")
plt.show()
```

4.3.3 Задание

Дана функция $|cos(𝑥𝑒^{(cos(𝑥)+ln(𝑥+1))})|$ Построить график на
интервале (0,10) с шагом 1 с заливкой площади и найти эту площадь под ним. Для вычисления площади воспользуйте функции trapz(),
вычисляющей интеграл по правилу трапеции.
Для ее корректной работы необходимо подключить следующие
библиотеки:

from scipy.integrate import simps

from numpy import trapz(не нужно импортировать)


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 10, 1)
y = np.abs(np.cos(x * np.exp(np.cos(x) + np.log(x + 1))))

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label='Функция')
plt.fill_between(x, y, alpha=0.3, color='blue', label='Площадь под кривой')
plt.xlabel('Значения по x см')
plt.ylabel('Значения по y см')
plt.title('График функции |cos(x * e^(cos(x) + ln(x+1)))|')
plt.legend()
plt.grid(True)
plt.show()

area = np.trapezoid(y, x)
print(f"Площадь под кривой: {area:.3f} см^2")
```

4.3.4 Задание

Откройте статистические данные по стоимости акций Apple,
Microsoft, Google за 12 месяцев 2021 г. Постройте 3 графика на
плоскости и оцените их динамику. Подумайте, как можно улучшить
читаемость результатов.
Для начальных данных допустимо использовать следующий ресурс:
https://finance.yahoo.com

```python
import yfinance as yf
import matplotlib.pyplot as plt

# 1. Загрузка данных
# Тикеры: AAPL (Apple), MSFT (Microsoft), GOOGL (Google)
tickers = ['AAPL', 'MSFT', 'GOOGL']

# Скачиваем данные за 2021 год с интервалом в 1 месяц (1mo)
data = yf.download(tickers, start='2021-01-01', end='2021-12-31', interval='1mo')

# Нас интересуют только цены закрытия месяца
close_prices = data['Close']

# 2. Построение графика
plt.figure(figsize=(10, 6))
plt.grid(True, linestyle='--', alpha=0.7)  # Сделали сетку чуть прозрачнее для комфортного чтения

# Используем close_prices.index (даты) для оси X
# Параметр marker='o' добавляет точки на каждый месяц, улучшая читаемость динамики
plt.plot(close_prices.index, close_prices['AAPL'], label="Apple", marker='o', linewidth=2)
plt.plot(close_prices.index, close_prices['MSFT'], label="Microsoft", marker='o', linewidth=2)
plt.plot(close_prices.index, close_prices['GOOGL'], label="Google", marker='o', linewidth=2)

plt.title("Стоимость акций Apple, Microsoft и Google за 2021 год")
plt.xlabel("Месяц")
plt.ylabel("Цена закрытия (USD)")
plt.legend()
plt.show()
```

### Анализ динамики и читаемости (Задание 4.3.4)

**Оценка динамики:**
1. **Общий тренд:** В течение 2021 года все три компании (Apple, Microsoft, Google) показали положительную динамику роста.
2. **Лидеры:** Акции Microsoft и Google демонстрировали более стабильный и крутой подъем по сравнению с Apple, которая имела периоды стагнации в середине года.
3. **Волатильность:** В конце года наблюдается повышенная волатильность (колебания цен), что характерно для технологического сектора в этот период.

**Улучшение читаемости:**
Для того чтобы графики были читабельными, в коде использовались следующие функции:
* **Сетка (plt.grid):** Позволяет точнее определить значение цены в конкретный месяц.
* **Маркеры ('o'):** Четко обозначают точки закрытия каждого месяца, исключая путаницу при изгибах линий.
* **Легенда (plt.legend):** Позволяет быстро сопоставить цвет линии с конкретной компанией.
* **Формат даты:** Использование `close_prices.index` автоматически подписывает месяцы на оси X.
* **Размер (figsize):** Увеличенное полотно предотвращает наложение подписей друг на друга.

4.3.5 Задание

Создайте простейший калькулятор, включающий основные действия
для двух переменных ′+′,′−′,′∙′,′÷′, а также вычисление
следующих функций: $e^{(x+y)},sin(x+y),cos(x+y),x^y$


```python
import math as m  # математические функции: sin, cos и константа e

print("Калькулятор. Чтобы выйти, введите 0 вместо операции.")

while True:
    x = float(input("Введите значение x (число): "))
    s = input("Введите операцию (+, -, *, /) или функцию (e, sin, cos, x**y), либо 0 для выхода: ")

    if s == "0":
        break

    if s in ("+", "-", "*", "/", "e", "sin", "cos", "x**y"):
        y = float(input("Введите значение y (число): "))

        if s == "+":
            print(f"Результат: сумма x и y = {x + y:.2f}")
        elif s == "-":
            print(f"Результат: разность x и y = {x - y:.2f}")
        elif s == "*":
            print(f"Результат: произведение x и y = {x * y:.2f}")
        elif s == "/":
            if y != 0:
                print(f"Результат: частное x / y = {x / y:.2f}")
            else:
                print("Ошибка: деление на ноль невозможно.")
        elif s == "e":
            print(f"Результат: e^(x + y) = {m.e ** (x + y):.2f}")
        elif s == "sin":
            print(f"Результат: sin(x + y) = {m.sin(x + y):.2f}")
        elif s == "cos":
            print(f"Результат: cos(x + y) = {m.cos(x + y):.2f}")
        elif s == "x**y":
            print(f"Результат: x в степени y (x**y) = {x ** y:.2f}")
    else:
        print("Неверный ввод операции/функции. Попробуйте ещё раз.")

print("Программа завершена.")
```
