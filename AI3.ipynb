### Теоретическая справка: Метрики расстояний
В данном задании мы вычисляем расстояние между объектами (точками). В машинном обучении расстояние показывает степень «похожести» объектов: чем меньше расстояние, тем больше объекты похожи друг на друга.

Мы используем три основные метрики:
1. **Евклидово расстояние** — геометрическое расстояние по прямой (кратчайший путь). Вычисляется через корень из суммы квадратов разностей координат.
2. **Квадрат евклидова расстояния** — та же метрика, но без извлечения корня. Используется алгоритмами для ускорения вычислений и более сильного «штрафования» за дальние выбросы.
3. **Хеммингово (Манхэттенское) расстояние** — сумма модулей разностей координат. Путь вычисляется только под прямым углом (как движение по кварталам города).
4. **Расстояние Чебышева** — берется максимальная разница только по одной, самой сильно отличающейся координате.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # Импорт для 3D-графики
import numpy as np # Импорт NumPy для числовых операций

# Создание 3D-графика
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

# Определение и отображение 4 точек в 3D-пространстве
point1 = ax.scatter(0,0,0, marker ="v")
point2 = ax.scatter(3,3,3, marker = "^")
point3 = ax.scatter(3,1,2, marker = "<")
point4 = ax.scatter(2,0,1, marker = ">")

# Координаты точек в виде массивов NumPy
x = np.array([0,0,0])
y = np.array([3,3,3])
z = np.array([3,1,2])
w = np.array([2,0,1])

plt.show() # Отображение 3D-графика

# Расчет и вывод расстояний между точками
# Используются L2-норма (Евклидово), квадрат L2-нормы, L-бесконечность (Чебышева) и L1-норма (Манхэттенское)

print("Расстояние между 1(v) и 2(^) точкой:")
print(f"{np.linalg.norm(x-y):.3f}") # L2-норма
print(f"{(np.linalg.norm(x-y) ** 2):.3f}") # Квадрат L2
print(f"{np.linalg.norm(x-y, ord = np.inf):.3f}") # L-бесконечность
print(f"{np.linalg.norm(x-y, ord=1):.3f}") # L1-норма

print("\nРасстояние между 1(v) и 3(<) точкой:")
print(f"{np.linalg.norm(x-z):.3f}") # Евклидово
print(f"{(np.linalg.norm(x-z) ** 2):.3f}") # Квадрат Евклида
print(f"{np.linalg.norm(x-z, ord = np.inf):.3f}") # Чебышева
print(f"{np.linalg.norm(x-z, ord=1):.3f}") # Манхэттенское

print("\nРасстояние между 1(v) и 4(>) точкой:")
print(f"{np.linalg.norm(x-w):.3f}") # Евклидово
print(f"{(np.linalg.norm(x-w) ** 2):.3f}") # Квадрат Евклида
print(f"{np.linalg.norm(x-w, ord = np.inf):.3f}") # Чебышева
print(f"{np.linalg.norm(x-w, ord=1):.3f}") # Манхэттенское

print("\nРасстояние между 2(^) и 3(<) точкой:")
print(f"{np.linalg.norm(y-z):.3f}") # Евклидово
print(f"{(np.linalg.norm(y-z) ** 2):.3f}") # Квадрат Евклида
print(f"{np.linalg.norm(y-z, ord = np.inf):.3f}") # Чебышева
print(f"{np.linalg.norm(y-z, ord=1):.3f}") # Манхэттенское

print("\nРасстояние между 2(^) и 4(>) точкой:")
print(f"{np.linalg.norm(y-w):.3f}") # Евклидово
print(f"{(np.linalg.norm(y-w) ** 2):.3f}") # Квадрат Евклида
print(f"{np.linalg.norm(y-w, ord = np.inf):.3f}") # Чебышева
print(f"{np.linalg.norm(y-w, ord=1):.3f}") # Манхэттенское

print("\nРасстояние между 3(^) и 4(>) точкой:")
print(f"{np.linalg.norm(w-z):.3f}") # Евклидово
print(f"{(np.linalg.norm(w-z) ** 2):.3f}") # Квадрат Евклида
print(f"{np.linalg.norm(w-z, ord = np.inf):.3f}") # Чебышева
print(f"{np.linalg.norm(w-z, ord=1):.3f}") # Манхэттенское"

#Так же можно использовать такия вариации если не знаем общей формулы расстояния Минковского
# 1. Евклидово расстояние (по прямой)
# В numpy можно использовать встроенную функцию linalg.norm
euclidean_dist = np.linalg.norm(p1 - p2)

# 3. Хеммингово (Манхэттенское) расстояние
# Считаем сумму модулей разностей по всем осям
manhattan_dist = np.sum(np.abs(p1 - p2))

# 4. Расстояние Чебышева
# Находим максимальное отклонение по одной из осей
chebyshev_dist = np.max(np.abs(p1 - p2))
```

2.3.1 Задание<br>
Для предыдущего примера поэкспериментируйте с параметрами
классификатора:
1. Установите другое количество ближайших соседей (k = 1, 5, 10).
2. Установите размер тестовой выборки 15% от всего датасета.
3. Постройте графики и оцените качество моделей, проанализируйте
результаты.

```python
from sklearn.model_selection import train_test_split # Инструмент для деления данных на обучение и проверку
from sklearn.neighbors import KNeighborsClassifier # Алгоритм 'k-ближайших соседей'
import seaborn as sns # Библиотека для красивых графиков и данных
import matplotlib.pyplot as plt
import numpy as np

# Загружаем готовый набор данных о цветах (ирисах)
iris = sns.load_dataset('iris')

# Делим данные: 85% отдаем модели на 'зубрежку', а 15% (test_size=0.15) оставляем для 'экзамена'
# x_train - характеристики цветов, y_train - их правильные названия
x_train, x_test, y_train, y_test = train_test_split(iris.iloc[:,:-1], iris.iloc[:,-1], test_size=0.15)

# Создаем 'мозг' модели. n_neighbors=10 значит, что при определении вида цветка
# программа будет смотреть на 10 самых похожих на него соседей
model = KNeighborsClassifier(n_neighbors=10)

# Обучаем модель на наших тренировочных данных
model.fit(x_train, y_train)

# Даем модели тестовые данные (которые она еще не видела) и просим угадать названия цветов
y_pred = model.predict(x_test)

# Настраиваем размер картинки для графика
plt.figure(figsize=(10,7))

# Рисуем все цветы из базы в виде точек. Цвет точки зависит от вида ириса
sns.scatterplot(data=iris, x='petal_width', y='petal_length', hue='species', s=70)
plt.grid() # Добавляем сетку для удобства

# Проверяем, где модель ошиблась:
for i in range(len(y_test)):
    # Если реальное название цветка не совпало с тем, что угадала модель...
    if np.array(y_test)[i] != y_pred[i]:
        # ...рисуем на этом месте большую красную точку (ошибку)
        plt.scatter(x_test.iloc[i,3], x_test.iloc[i,2], color='red', s=150)

from sklearn.metrics import accuracy_score
# Считаем процент правильных ответов (точность) и выводим на экран
print(f'accuracy (точность): {accuracy_score(y_test, y_pred):.3f}')
```
